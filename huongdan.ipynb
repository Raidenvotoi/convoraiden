```python
import pandas as pd

housing_path = "https://media.githubusercontent.com/media/tiepvupsu/tabml_data/master/california_housing/"
df_housing = pd.read_csv(housing_path + "housing.csv")
df_housing.head()
```

```python
df_housing.describe()
```

```python
df_housing["ocean_proximity"].value_counts()
```

```python
from matplotlib import pyplot as plt

df_housing.hist(bins=50, figsize=(20, 15))
```

```python
from pandas.plotting import scatter_matrix
import numpy as np
columns =df_housing.select_dtypes(include=[np.number]).columns
scatter_matrix(df_housing[columns], figsize=(12,10), hist_kwds={'bins': 50})
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

correlation_matrix = df_housing.select_dtypes(include=[np.number]).corr()
plt.figure(figsize=(15, 10))
sns.heatmap(correlation_matrix, annot=True, linewidths=0.5, fmt='.2f',)
```

```python
plt.boxplot(df_housing['median_house_value'])

# Add labels and title
plt.xlabel('num of PRICE')
plt.ylabel('Frequency')
plt.title('PRICE')

# Show the plot
plt.show()
```

```python
df_housing.plot(
    kind="scatter",
    x="longitude",
    y="latitude",
    alpha=0.5,
    s=df_housing["population"]/1000,
    label="population",
    figsize=(10, 8),
    c="median_house_value",
    cmap=plt.get_cmap("jet"),
    colorbar=True
)
plt.legend()
```

```python
x_label = "median_house_value"
y_label = "median_income"
plt.figure(figsize=(12, 12))
plt.scatter(df_housing[x_label], df_housing[y_label], s=4)
plt.xlabel(x_label, fontsize=18)
plt.ylabel(y_label, fontsize=18);
```
